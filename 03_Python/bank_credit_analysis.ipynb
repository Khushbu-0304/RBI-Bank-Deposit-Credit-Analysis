```python
import pandas as pd

df=pd.read_excel(
    "/content/Bank_Credit_Deposit_Analysis.xlsx",
    sheet_name="Master_Data"
)
df.head()
```

```python
df.columns = [
    "State",
    "Deposit_2020", "Deposit_2021", "Deposit_2022", "Deposit_2023", "Deposit_2024", "Deposit_2025",
    "Credit_2020", "Credit_2021", "Credit_2022", "Credit_2023", "Credit_2024", "Credit_2025"
]

df.head()
```

```python
# keep only valid state rows
df = df[df["State"].notna()]
df = df[df["State"].astype(str).str.strip() != ""]
df = df[df["State"].astype(str).str.strip() != "-1"]

df.reset_index(drop=True, inplace=True)
df.head()
```

```python
deposit_long = df.melt(
    id_vars="State",
    value_vars=["Deposit_2020", "Deposit_2021", "Deposit_2022", "Deposit_2023", "Deposit_2024", "Deposit_2025"],
    var_name="Year",
    value_name="Deposit"
)

credit_long = df.melt(
    id_vars="State",
    value_vars=["Credit_2020", "Credit_2021", "Credit_2022", "Credit_2023", "Credit_2024", "Credit_2025"],
    var_name="Year",
    value_name="Credit"
)
```

```python
deposit_long["Year"] = deposit_long["Year"].str.replace("Deposit_", "")
credit_long["Year"] = credit_long["Year"].str.replace("Credit_", "")
```

```python
banking_long = pd.merge(
    deposit_long,
    credit_long,
    on=["State", "Year"]
)

banking_long.head(10)
```

```python
banking_long.shape
```

```python
# remove blank state rows if any
banking_long = banking_long[banking_long["State"].notna()]

# remove rows where State is empty string
banking_long = banking_long[banking_long["State"].astype(str).str.strip() != ""]

# clean Year column and convert to integer
banking_long["Year"] = banking_long["Year"].astype(int)

# convert Deposit and Credit to numeric
banking_long["Deposit"] = pd.to_numeric(banking_long["Deposit"], errors="coerce")
banking_long["Credit"] = pd.to_numeric(banking_long["Credit"], errors="coerce")

# drop rows where Deposit or Credit is missing
banking_long = banking_long.dropna(subset=["Deposit", "Credit"])

# reset index
banking_long = banking_long.reset_index(drop=True)

banking_long.head()
```

```python
region_map = {
    "Haryana": "Northern",
    "Himachal Pradesh": "Northern",
    "Jammu & Kashmir": "Northern",
    "Punjab": "Northern",
    "Rajasthan": "Northern",
    "Chandigarh": "Northern",
    "Delhi": "Northern",
    "Ladakh": "Northern",

    "Arunachal Pradesh": "North-Eastern",
    "Assam": "North-Eastern",
    "Manipur": "North-Eastern",
    "Meghalaya": "North-Eastern",
    "Mizoram": "North-Eastern",
    "Nagaland": "North-Eastern",
    "Tripura": "North-Eastern",

    "Bihar": "Eastern",
    "Jharkhand": "Eastern",
    "Odisha": "Eastern",
    "Sikkim": "Eastern",
    "West Bengal": "Eastern",
    "Andaman & Nicobar Islands": "Eastern",

    "Chhattisgarh": "Central",
    "Madhya Pradesh": "Central",
    "Uttar Pradesh": "Central",
    "Uttarakhand": "Central",

    "Goa": "Western",
    "Gujarat": "Western",
    "Maharashtra": "Western",
    "Dadra & Nagar Haveli": "Western",

    "Andhra Pradesh": "Southern",
    "Karnataka": "Southern",
    "Kerala": "Southern",
    "Tamil Nadu": "Southern",
    "Telangana": "Southern",
    "Puducherry": "Southern",
    "Lakshadweep": "Southern"
}

banking_long["Region"] = banking_long["State"].map(region_map)

banking_long.head()
```

```python
banking_long = banking_long[["Year", "Region", "State", "Deposit", "Credit"]]
banking_long.head()
```
